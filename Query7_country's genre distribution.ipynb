```python
import pandas as pd
import matplotlib.pyplot as plt

add="C:/Users/YUNUS/Downloads/data-1556134822348.csv"
df=pd.read_csv(add,sep=',',header=0,quotechar='"')

f = plt.figure(figsize=(20,20))

i=1
for country in df["BillingCountry"].value_counts().index:
    
    ax = f.add_subplot(3,2,i)
    ax.set_title("{}'s genre distribution".format(country))
    ax.barh(df.loc[df.BillingCountry==country,'Name'],df.loc[df.BillingCountry==country,'sum'])
    i+=1
plt.show()
```
